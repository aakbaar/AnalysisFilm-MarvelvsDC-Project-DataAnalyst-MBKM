```python
df_update = df.copy()
df_update['IMDB_Score'] = df_update['IMDB_Score'].fillna(df_update['IMDB_Score'].median())
df_update['Votes'] = df_update['Votes'].fillna(df_update['Votes'].median())
df_update['Rating'] = df_update['Rating'].fillna(df_update['Rating'].mode()[0])
df_update['Genre'] = df_update['Genre'].fillna(df_update['Genre'].mode()[0])
df_update['Director'] = df_update['Director'].fillna(df_update['Director'].mode()[0])
df_update['Actor'] = df_update['Actor'].fillna(df_update['Actor'].mode()[0])
df_update.head()
```

```python
df_update.info()
```

```python
def f(columns):
    if columns['IMDB_Score']>=8.5:
        return 'Very Good'
    elif columns['IMDB_Score']>=7.0:
        return 'Good'
    else:
        return 'Bad'
```

```python
df_update['Value_IMDB'] = df.apply(f,axis=1)
df_update
```

```python
df_update.to_csv(r'marvel_fix.csv',index = False)
```
